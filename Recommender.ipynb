```python
import pandas as pd
  
# Creating the dataframe 
ratings = pd.read_csv('ContentBasedRecommenderSystem/ratings.csv', usecols=['userId','movieId','rating'])
movies = pd.read_csv('ContentBasedRecommenderSystem/movies.csv', usecols=['movieId','title'])
ratings_movies = pd.merge(ratings, movies, how='inner', on='movieId')
dfv1 = ratings_movies.pivot_table(index='title',columns='userId',values='rating')
dfv1 = dfv1.iloc[:,:200]#choose only 200 users
dfv1 = dfv1.dropna(thresh=3, axis=0)#remove any movies rated less than 3 times to reduce noise
# we used 2 variables here, dfv1 and df, because we will use the dataframe with the nans when we check whether a user watched a movie or not
df=dfv1.fillna(0,axis=0)#fill nan with zero
df.head()

```

```python
from sklearn.metrics.pairwise import cosine_similarity

def standardize(column): #due to the NAN to zero replacement, we want to standarize the ratings to bring the mean to zero
    new_column = (column - column.mean())/(column.max()-column.min())
    return new_column

df_std = df.apply(standardize)

corrMatrix = pd.DataFrame(cosine_similarity(df_std),index=df.index,columns=df.index)
corrMatrix.head()
```

```python
def get_similar_given_movie(movie_name,number_of_movies):## takes movie name and the number of similar movies we want to return
    similarMovies=corrMatrix[movie_name].sort_values(ascending=False).index.to_list()
    if number_of_movies<len(similarMovies):
        return similarMovies[1:number_of_movies+1]# the first movie in this list will be the same movie that we passed to the function, so we exclude it
    else:
         return similarMovies[1:]



def get_similar_given_user(id_user):

    topRatedMovies=dfv1.iloc[:,id_user-1].dropna(axis=0).sort_values(ascending=False).index.to_list() #names of the  top rated ""watched""" movies by the user,,, 
    #(id_user-1) due to zero indexing,,,
    #dropna is used to drop the movies that the user didn't watch,    #note that we used (dfv1) , that has nan values#

    recommended_movies=[]


    #To recommend 3 movies to a user
    # we have 2 options, either we choose one movie the user watched and liked most, then recommend 3 movies similar to it , 
    # or we choose 3 movies the user liked, and get one similar movie to each one of them
    #but we should make sure that the user didn't watch the movies that we will recommnend


    # we chose the 2nd option, to get 3 movies the user liked, and get one similar movie to each one of them
    num_movies_recommended=3
    for i in range (num_movies_recommended) : 
        movies=get_similar_given_movie(movie_name=topRatedMovies[i],number_of_movies=20)# get 20 movies similar to the ith movie that the user liked

        for movie in movies: 
            
            if movie not in topRatedMovies: #if the movie we found hasn't been watched by the user yet
                recommended_movies.append(movie)#then, recommend it
                break
            ##we assumed that we will find a movie the user didn't watch in the 20 movie that we got, so no else statement

    return recommended_movies
            

```

```python
print(get_similar_given_movie(movie_name='Toy Story (1995)',number_of_movies=10),"\n")



print(get_similar_given_movie(movie_name='Waiting to Exhale (1995)',number_of_movies=10),"\n")


print(get_similar_given_user(id_user=200))

```
